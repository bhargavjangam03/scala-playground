```scala
import org.apache.spark.sql.SparkSession
import org.apache.spark.sql.functions._
import org.apache.spark.sql.types._

// Initialize SparkSession
val spark = SparkSession.builder()
  .appName("Time-Based Data Partitioning for Ratings")
  .getOrCreate()
```

```scala
val ratingsPath = "gs://bhargav-assignments/Day16And17Task/rating.csv"
val outputPath = "hdfs:///user/Day16AndDay17/case_study_5/ratings"

// Load ratings.csv as a DataFrame
val ratingsDF = spark.read
  .option("header", "true")
  .csv(ratingsPath)

//Show the structure of the DataFrame
println("Ratings DataFrame:")
ratingsDF.show(5, truncate = false)
```

```scala
val ratingsWithYearDF = ratingsDF.withColumn("year", year(col("timestamp"))) 

//Show the DataFrame with the year column
println("Ratings DataFrame with Year:")
ratingsWithYearDF.show(5, truncate = false)

```

```scala
// Save the ratings DataFrame partitioned by year in Parquet format to HDFS
ratingsWithYearDF.write
  .mode("overwrite")
  .partitionBy("year")  // Partition the data by year
  .parquet(outputPath)

// Saved data to HDFS
println(s"Ratings data saved to $outputPath partitioned by year.")
```

```scala
// Stop the SparkSession
spark.stop()
```
